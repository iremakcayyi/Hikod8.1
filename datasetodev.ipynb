```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df.review_scores_rating, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Review Scores Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Review Scores Rating')
plt.show()
```

```python
room_type_counts = df['room_type'].value_counts()
room_type_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Number of Listings by Room Type')
plt.show()
```

```python
plt.scatter(df['longitude'], df['latitude'], color='orange', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Listings')
plt.show()
```

```python
avg_price_by_property_type = df.groupby('property_type')['log_price'].mean()
print(avg_price_by_property_type)
```
